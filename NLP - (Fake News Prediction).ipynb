# NLP STEPS

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
stem_port = PorterStemmer()
```

```python
import nltk
nltk.download('stopwords')
```

```python
def stemming(content):
  stemmed_content = re.sub('[^a-zA-Z]',' ',content)
  stemmed_content = stemmed_content.lower()
  stemmed_content = stemmed_content.split()
  stemmed_content = [stem_port.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
  stemmed_content = ' '.join(stemmed_content)
  return stemmed_content
```

```python
df['text'] = df['text'].apply(stemming)
```

```python
df.head()
```

```python
x = df['text'].values
y = df['label'].values
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
vectorizer.fit(x)
x = vectorizer.transform(x)
```

```python
print(x)
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,stratify=y,random_state=2)    # Over Sampling Data Balancing  # The stratify = y parameter is used in machine learning, particularly in data splitting and cross validation, to ensure that the distribution of classes in the training and testing sets (or folds) matches the distribution in the entire data set. This is particularly important in classification problems where the class distribution might be imbalanced.
```

**MODEL BUILDING**

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
accuracy
```
